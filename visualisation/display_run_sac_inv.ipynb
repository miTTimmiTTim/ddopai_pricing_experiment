```python
api = wandb.Api()
```

```python
entity = 'timlachner' 
project = 'pricing_cMDP_test'
run_ids = ["k2rhtgpj"]
run_names = ["SAC"]
```

```python
runs = []
for run_id in run_ids:
    runs.append(api.run(f"{entity}/{project}/{run_id}"))
```

```python
sac = runs[0].history()
```

```python
sac.columns
```

```python

sac_filtered = sac[["Epoch", "t", "True_Cumulative_Reward", "Inventory"]]
```

```python
sac_filtered['Epoch'] = sac_filtered['Epoch'].astype(str)
```

```python
from lets_plot import *

LetsPlot.setup_html()

# Plotting
plot = (
    ggplot(sac_filtered, aes(x='t'))
    + geom_line(aes(y='True_Cumulative_Reward', color='Epoch'), size=1.2)
    + geom_line(aes(y='Inventory', color='Epoch'), linetype='dashed', size=1.2)
    + labs(x='T', y='Cumulative Reward/Inventory', title='Performance of Continous RL on MDPs with Inventory Constraints')
    + theme_minimal()
    + guides(color=guide_legend(title='Epoch'))
    + ggsize(1000, 600)
)

plot.show()
```

```python
ggsave(plot, filename='plot.svg')
```
